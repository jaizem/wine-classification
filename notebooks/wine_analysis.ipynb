# Wine Categorization with Clustering 

```python
import sys
sys.path.append("../")  # adds the parent directory to Python path

from src.data_preprocessing import load_data, clean_data, scale_features
from src.clustering_models import train_kmeans, train_agglomerative
from src.evaluation import silhouette_score_eval, davies_bouldin_eval, silhouette_vs_k
from src.visualization import plot_pca_clusters, plot_heatmap, silhouette_plot, pca_transform_and_plot
```

### Load and preprocess data

```python
wine = load_data("../data/wine.csv")
wine_cleaned = clean_data(wine)
X_scaled = scale_features(wine_cleaned)
```

### Train models

```python
kmeans_model, kmeans_labels = train_kmeans(X_scaled)
agg_model, agg_labels = train_agglomerative(X_scaled)
```

### Evaluate

```python
sil_score = silhouette_score_eval(X_scaled, kmeans_labels)
db_score = davies_bouldin_eval(X_scaled, kmeans_labels)

scores = silhouette_vs_k(X_scaled, k_range=range(2, 11), plot=True)

print("KMeans scores:", scores["kmeans"])
print("Agglomerative scores:", scores["agg"])
```

### Visualizations

```python
# Run PCA and plot heatmap
X_pca, loadings = pca_transform_and_plot(X_scaled, n_components=3, plot=True)

# Train k-means
kmeans_model, kmeans_labels = train_kmeans(X_scaled)

# Plot clusters
plot_pca_clusters(X_scaled, kmeans_labels, title="KMeans Clusters")

# Plot silhouette
silhouette_plot(X_scaled, kmeans_labels)
```
